```python
import os
import urllib.request
import datetime

print("Start Program")

base_url = ['https://s3.amazonaws.com/nyc-tlc/trip+data/yellow_tripdata_',
            'https://s3.amazonaws.com/nyc-tlc/trip+data/green_tripdata_',
            'https://nyc-tlc.s3.amazonaws.com/trip+data/fhv_tripdata_',
            'https://nyc-tlc.s3.amazonaws.com/trip+data/fhvhv_tripdata_']

filenames = ["tf_yellow_tripdata", "tf_green_tripdata", "tf_fhv_tripdata",  "tf_fhvhv_tripdata"]


def appendData(year, month):
    
    for i in range(4):
        month_str = '{0:02d}'.format(month)
        year_month_str = str(year) + "-" + month_str + ".csv"
        try:
            url = base_url[i] + year_month_str
            print(url)

            datasource = urllib.request.urlopen(url)

            merge_file = filenames[i] + ".csv"
            new_file = True
            if not os.path.isfile(merge_file):
                fp = open(merge_file, "w")
                new_file = True
            else:  # Else it exists so append without writing the header (to prevent overwriting data within source file)
                fp = open(merge_file, "a")
                new_file = False

            count = 0
            while True:
                line = datasource.readline()
                if new_file == False and count == 0:    # If downloaded file exists skip header as you already have it from previous file 
                    count += 1
                    continue

                if not line: # check if csv stream is ended or not
                    break

                line = line.decode("utf-8").rstrip() # Convert binary value to string and remove the last return character('\r\n')

                if not line: # skip empty line
                    continue

                line = line + '\n'
                fp.write(line)
                count += 1

                if count % 50000 == 0:
                    print("Line " + str(count) + " is processed")

                if count > 100: # just save 1000 records for each csv file for quick testing
                    break

            fp.close()
        except Exception as e:
            print("An Exception URL", str(e))

    print("Year ", str(year) , "Month =", str(month))
```

```python
print("Whole Data")

# get currrent year, moonth
today = datetime.datetime.now()
print("Today ", today.year, today.month, today.day)

# extract all history csv data and merge except for this month
stop_flag = False
for year in range(2009, today.year + 1):
    if stop_flag == True:
        break
        
    for month in range(1, 13):
        if year == today.year and month == today.month:
            stop_flag = True
            break
            
        appendData(year, month)
    
```

```python
# This is delta script for this month, you need to run this script for each month after collectin g whole history
appendData(today.year, today.month)
```
