```python
# Testing url loader
import os
import urllib.request

print("Start Program")

base_url = ['https://s3.amazonaws.com/nyc-tlc/trip+data/yellow_tripdata_',
            'https://s3.amazonaws.com/nyc-tlc/trip+data/green_tripdata_',
            'https://nyc-tlc.s3.amazonaws.com/trip+data/fhv_tripdata_',
            'https://nyc-tlc.s3.amazonaws.com/trip+data/fhvhv_tripdata_']

filenames = ["yellow_tripdata", "green_tripdata", "fhv_tripdata",  "fhvhv_tripdata"]

year_beginning = 2009
year_ending = 2021

for i in range(4):
    for year in range(year_beginning, year_ending):
        for month in range(1, 13):
            month_str = '{0:02d}'.format(month)
            year_month_str = str(year) + "-" + month_str + ".csv"
            try:
                url = base_url[i] + year_month_str
                print(url)
            except Exception as e:
                print("An Exception URL", str(e))

print("End Program")
```
